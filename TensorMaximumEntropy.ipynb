## 1 Quantifying primary features across different modes of the data 

Each dataset forms a tensor $X \in \mathbb{R}^{T \times N \times C}$. To quantify the primary temporal, neural and condition
features, we calculated the marginal mean and covariance across each of these three modes. Regarding the mean, we followed standard practice and, without, loss of generality, centered the data to form a tensor $\bar{X} \in \mathbb{R}^{T \times N \times C}$ with zero mean across the temporal mode, neural mode and condition mode.

### Step 1.1 Compute $M$ & define zero-mean data $\bar{X}$

TME shuffles the data while keeping the mean and the covariance the same. First, we compute the mean $M \in \mathbb{R}^{T \times N \times C}$ (as in the article) such that 
\begin{align}
    & \frac{1}{NC} \sum_{n=1}^N \sum_{c=1}^C X(:, n, c) - M(:, n, c) = \mathbf{0} \\
    & \frac{1}{NT} \sum_{n=1}^N \sum_{t=1}^T X(t, n, :) - M(t, n, :) = \mathbf{0} \\
    & \frac{1}{CT} \sum_{c=1}^C \sum_{t=1}^T X(t, :, c) - M(t, :, c) = \mathbf{0} \\
\end{align}

so

\begin{align}
    M_T &= \frac{1}{NC} \sum_{n=1}^N \sum_{c=1}^C X(:, n, c) \\
    M_N &= \frac{1}{NC} \sum_{n=1}^N \sum_{c=1}^C X(:, n, c) \\
    M_C &= \frac{1}{NC} \sum_{n=1}^N \sum_{c=1}^C X(:, n, c)
\end{align}

We can now use the computed means to get the demeaned dataset 
\begin{align}
    \bar{X} = X - M
\end{align}

```python
''' Get data dimensions '''
X0 = X

''' Compute mean along dimension T '''
MT = np.mean(X0, axis=(1, 2))
X0 = X0 - (MT.reshape((T, 1)) @ np.ones(shape=(1, N))).reshape((T, N, 1)) @ np.ones((1, C))

''' Compute mean along dimension N '''
MN = np.mean(X0, axis=(0, 2))
X0 = X0 - (np.ones(shape=(T, 1)) @ MN.reshape((1, N))).reshape((T, N, 1)) @ np.ones((1, C))

''' Compute mean along dimension C '''
MC = np.mean(X0, axis=(0, 1))
X0 = X0 - (np.ones(shape=(T, 1)) @ np.ones(shape=(1, N))).reshape((T, N, 1)) @ MC.reshape((1, C))

''' Compute demeaned data'''
X_bar = X0
M = X - X_bar
```

### Step 1.2 Compute covariance matrices

With the zero-mean dataset, we calculate the covariance matrices across times, neurons and conditions
\begin{align}
    \Sigma_T &= \sum_{n=1}^N \sum_{c=1}^C \bar{X}(:, n, c) \bar{X}(:, n, c)^\top \\
    \Sigma_N &= \sum_{t=1}^T \sum_{c=1}^C \bar{X}(t, :, c) \bar{X}(t, :, c)^\top \\
    \Sigma_C &= \sum_{n=1}^N \sum_{t=1}^T \bar{X}(t, n, :) \bar{X}(t, n, :)^\top \\
\end{align}

The marginal mean tensor and covariance matrices quantify the basic univariate and pairwise structure of the data across each of the temporal, neural and condition modes.
